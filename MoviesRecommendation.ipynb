****Importing Libraries For NLP And Apply NLP:****

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

**NLP Data Preprocess:**

1. **Removing Punctuations:**|\/#$%+-=
2. **Removing Stopwords:** a , the, of , when, why, are , my etc.
3. **Lemmatization:** every word in its root form.eg= do, done, did, doing they all are do
4. **Lowercase the words:**


```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
def preprocess_text(text):
  #lowercase
  text = str(text).lower()
  #removed the punctuations
  text = re.sub(r'[^a-zA-Z\s]', '', text)

  words = text.split()
  #remove stopwords
  words = [word for word in words if word not in stop_words]

  #lemmatize
  words = [lemmatizer.lemmatize(word) for word in words]

  return " ".join(words)
```

```python
df['tags'] = df['tags'].apply(preprocess_text)
```

```python
df['tags'][0]
```

**When We drop any data Index number might shift so we have to reset the Index**

```python
df = df.reset_index(drop = True)
```

```python
indices = pd.Series(df.index,index = df['title']).drop_duplicates()
indices
```

**Converting Text into Numbers or Vectors:**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(max_features=50000,ngram_range=(1,2),stop_words='english')
```

```python
tfidf_matrix = tfidf.fit_transform(df['tags'])
```

```python
tfidf_matrix
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

**Using Cosine Similarity:** using cos of the vectors finding the pattern for the pridiction.






```python
def recommend(title, n = 10):
  if title not in indices:
    return ['Movie not found']

  idx = indices[title]
  sim_score = cosine_similarity(tfidf_matrix[idx], tfidf_matrix).flatten()
  similar_idx = sim_score.argsort()[::-1][1:n+1] #argsort saperate in the basis of index or column
  return df['title'].iloc[similar_idx]
```

**Testing The Model with some Examples:**

```python
recommend('Avatar',5)
```

```python
recommend('Toy Story')
```

```python
recommend('Jumanji',7)
```
